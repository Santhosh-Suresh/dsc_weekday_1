```python
x = pd.Series(np.arange(10))
y = x*2
y1 = x**2

plt.figure(figsize=(15,5))
plt.plot(x,y, color='c')
plt.plot(x,y1)
plt.show()
```

```python
x = np.arange(10)
y = x*2
y1 = x**2

print(x)

plt.figure(figsize=(15,5))
plt.plot(x,y, color='c')
plt.plot(x,y1)
plt.xticks(x)
plt.yticks(y1)
plt.title("Title")
plt.xlabel("time")
plt.ylabel("cases")
plt.show()
```

```python
x = pd.Series(np.arange(10))
y = x*2
y1 = x**2

plt.figure(figsize=(15,5))
plt.plot(x,y, color='c', marker='.')
plt.plot(x,y1)
plt.show()
```

```python
x = pd.Series(np.arange(10))
y = x*2
y1 = x**2

plt.figure(figsize=(15,5))
plt.plot(x,y, "r^")
plt.show()
```
