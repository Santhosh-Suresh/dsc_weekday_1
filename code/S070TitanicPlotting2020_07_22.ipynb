
### Titanic Plotting

- plot a horizontal bar-chart depicting the ratio of males to females
- plot a bar-chart depicting the ratio of male-survived, female-survived, male-dead, female-dead.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("titanic_filtered.csv")
print(df.head())
```

```python
# plot histogram of ages of people
plt.figure(figsize=(16,6))
df.age.plot(kind="hist", bins=40, rwidth=0.9)
plt.show()
```

```python
# plot no of people in each pclass on bar chart
s = df.pclass.value_counts()
print(s)
s.plot(kind="bar")
plt.show()

a = df.groupby('pclass').size()
print(a)
a.plot.bar()
plt.show()

s = df.pclass.value_counts()
print(s)
s.plot(kind="barh")
plt.show()
```

```python
# plot a bar-chart depicting the ratio of male-survived, female-survived, male-dead, female-dead.
bc=df[['gender', 'survived']]
dc=df.groupby(['gender', 'survived'])
cd=dc.size()
cd.plot.bar()
plt.show()
```

```python
t = pd.DataFrame({"A":[10,20], "B":[15,30]})
print(t)

# 
t.plot(kind="bar")
plt.show()


t.plot(kind="bar", stacked=True)
plt.show()


plt.bar(t.index, t.A.values)
plt.bar(t.index, t.B.values, bottom=t.A.values)
plt.xticks(t.index)
plt.show()
```

```python
# plot a stacked bar chart of number of male-survived,dead and female-survived, dead.
grp = df.groupby(by=["gender", "survived"])
s = grp.size()
print()
print(s)

print()
print(s.index)

print()
res = s.unstack()
print(res)
res.plot(kind="bar", stacked=True)
plt.show()


res.plot(kind="bar")
plt.show()
```

```python
# plot a stacked bar chart of number of male-survived,dead and female-survived, dead.
grp = df.groupby(by=["gender", "survived", "pclass"])
s = grp.size()
print()
print(s)

print()
print(s.index)

print()
res = s.unstack()
print(res)
res.plot(kind="bar", stacked=True)
```

```python
# plot a stacked bar chart of number of male-survived,dead and female-survived, dead.
grp = df.groupby(by=["gender", "survived", "pclass"])
s = grp.size()
print()
print(s)

print()
res = s.unstack()
print(res)

print()
idx = res.index.to_flat_index()
print(idx)

print()
res.index = idx
print(res)

res.plot(kind="bar", stacked=True)
```

```python
# plot a stacked bar chart of number of male-survived,dead and female-survived, dead.
grp = df.groupby(by=["gender", "survived", "pclass"])
s = grp.size()
print()
print(s)

print()
res = s.unstack()
print(res)

print()
res = s.unstack(level=0)
print(res)

print()
res = s.unstack(level=1)
print(res)

print()
res = s.unstack(level=2)
print(res)
```

```python
# plot box plot of ages of survived vs not-survived
c=df.groupby(by=['survived'])
print(c.age)

g1 = c.get_group(0).age
g2 = c.get_group(1).age

print(len(g1), len(g2))
plt.boxplot([g1, g2])
plt.show()
```

```python
# plot scatter plot of ages vs survived.
plt.figure(figsize=(15,5))

for a, s in zip(df.age, df.survived):
    plt.scatter(a, s, alpha=0.5, s=a, color='b')
plt.show()
```
