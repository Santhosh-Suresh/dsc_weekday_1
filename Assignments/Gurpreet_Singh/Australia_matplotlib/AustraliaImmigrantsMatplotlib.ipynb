```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Use AustraliaFiltered.xlsx dataset generated earlier.
df = pd.read_excel('AustraliaFiltered.xlsx')
df.head()
```

```python
dict1 = {"Name":["Raja","Lucky","Raja"] ,"Area":[1,2,3],"planets":["Mars","Mars","Juputer"] , "grade":["A","B","C"]  }
dummy = pd.DataFrame(dict1)
print(dummy.groupby("Name").count())
```

```python
# Plot bar chart of total no of immigrants for countires top 5 countires having max no of total immigrants
top  = df.nlargest(5, 'Total')
# print(top.columns)
# print(top.shape)
# print(top[['Country','Total']]) 
# print(df[['Country','Total']])
plt.figure(figsize=(20,6))
plt.bar(top.Country, top.Total , color = 'y')
```

```python
# Find total immigrants from each category by area name and plot barchart for each area

plt.figure(figsize=(15,5))
fg = df.groupby("AreaName")["Total"].sum().plot(kind = 'bar')
plt.show()
```

```python
# Plot total immigration trend(immigration of world over the years i.e. year vs Total count for that year) using line plot.
years = df.columns[4:len(df.columns)-1]
year_data =[]
for i in years:
    year_data.append(df[i].sum(axis = 0))
print(years , year_data)
plt.figure(figsize=(16,5))
plt.plot(years, year_data)
plt.xlabel('Years')
plt.ylabel('Immigrants')
plt.title('Immigrants year-wise')
plt.grid() #optional
plt.show()
```

```python
# Compare Immigration Trend for Asia, Africa and Europe AreaName (Use Line plot).
years = df.columns[4:len(df.columns)-1]
asia_data = []
africa_data = []
europe_data = []
for i in years:
    asia_data.append(df[df['AreaName']=="Asia"][i].sum(axis = 0))
    africa_data.append(df[df['AreaName']=="Africa"][i].sum(axis = 0))
    europe_data.append(df[df['AreaName']=="Europe"][i].sum(axis = 0))

        

plt.figure(figsize=(16,5))
plt.plot(years, asia_data,label = 'Asia')
plt.plot(years,africa_data,label = 'Africa')
plt.plot(years, europe_data,label = 'Europe')
plt.grid()  
plt.xlabel('Years')
plt.ylabel('Total Immigrants')
plt.title('Asia, Africa Europe Immigrants year-wise')
plt.legend()
plt.show()
```

```python
# Plot Total Immigrants based on DevelopmenRegion :
plt.figure(figsize=(16,5))
print(df.groupby("DevName")["Total"].sum())
df.groupby("DevName")["Total"].sum().plot(kind = 'bar')
plt.grid()  
plt.title('Total Immigrants based on DevelopmenRegion')
plt.show()
```

```python
# Plot immigration trend for India using line plot
years = df.columns[4:len(df.columns)-1]
india_data = []
for i in years:
    india_data.append(df[df['Country']=="India"][i].sum(axis = 0))

print(india_data)        

plt.figure(figsize=(16,5))
plt.plot(years, india_data,label = 'India')
plt.grid()  
plt.xlabel('Years')
plt.ylabel('Total Immigrants')
plt.title('India Immigrants year-wise')
plt.show()
```

```python
# Plot immigration Trend(year vs count) for the coutries with top 5 total immigrants. Use Line plot
# nlargest(5,df.groupby("Country")["Total"].sum())
print(len(df) , len(df['Country'].unique()))
top_5 = df.nlargest(5, 'Total')

plt.figure(figsize = (16,6))
for top in top_5.values:
    country_name = top[0]
    year_count = top[4:33]
    plt.plot(years, year_count,label = country_name)
plt.title('Top 5 countries total immigrant count')
plt.xlabel('year')
plt.ylabel('Total Immigrants')
plt.legend(loc=2)
plt.show()
```

```python
# Plot immigration Trend(year vs count) for the coutries with top 5 total immigrants.
# For the above use area plot with alpha = 0.5
years = df.columns[4:len(df.columns)-1]
top_5 = df.nlargest(5, 'Total')['Country']
plt.figure(figsize = (16,6))
for i in top_5.values:
    country_wise_sum = df[df.Country == i][df.columns[4:len(df.columns)-1]].sum()
    years = country_wise_sum.index.astype(str)
    year_count = country_wise_sum.values
    plt.fill_between(years, year_count,label=i,alpha=0.5)
plt.title('Top 5 countries total immigrant count')
plt.xlabel('year')
plt.ylabel('Total Immigrants')
plt.legend(loc=2)
plt.show()
```

```python
# Compare Immigration Trend for top-5 countires using stacked area plot
years = df.columns[4:len(df.columns)-1]
top_5 = df.nlargest(5, 'Total')['Country']
plt.figure(figsize = (16,6))

for i in top_5.values:
    country_wise_sum = df[df.Country == i][df.columns[4:len(df.columns)-1]].sum()
    years = country_wise_sum.index.astype(str)
    year_count = country_wise_sum.values
    plt.fill_between(years, year_count,label=i,alpha=0.5)
plt.plot(kind ="area", stacked=True, figsize=(20,5))
plt.legend(loc=2)

plt.show()    
```
