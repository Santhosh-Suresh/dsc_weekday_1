```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
```

```python
xls = pd.ExcelFile('AustraliaFiltered.xlsx')
d = xls.parse('Sheet1')
print(d.columns)
print(d.index)
```

```python
a=d.nlargest(5,columns='Total')
plt.figure(figsize=(20,5))
plt.bar(a.Country,a.Total)
```

```python
d.replace('..',0,inplace=True)
```

```python
x=d[d.Country=='India'].columns[-35:-1]
y=d[d.Country=='India'].values[0][-35:-1]
plt.figure(figsize=(20,5))
plt.plot(x,y)
plt.show()
```

```python
a=d.nlargest(5,columns='Total')
x=a.columns[-35:-1]
y1=a.values[0][-35:-1]
y2=a.values[1][-35:-1]
y3=a.values[2][-35:-1]
y4=a.values[3][-35:-1]
y5=a.values[4][-35:-1]
plt.figure(figsize=(20,10))
plt.plot(x,y1,label=a.Country.iloc[0])
plt.plot(x,y2,label=a.Country.iloc[1])
plt.plot(x,y3,label=a.Country.iloc[2])
plt.plot(x,y4,label=a.Country.iloc[3])
plt.plot(x,y5,label=a.Country.iloc[4])
plt.legend()
plt.show()
```

```python
a=d.groupby(by='DevName').groups
y1=d.iloc[a['Developed regions']].Total.values.sum()
y2=d.iloc[a['Developing regions']].Total.values.sum()
x1='Developed Region'
x2='Developing Region'
plt.figure(figsize=(5,5))
plt.bar(x1,y1,width=0.2)
plt.bar(x2,y2,width=0.2)
plt.show()
```

```python
a=d.groupby(by='AreaName').sum()
plt.figure(figsize=(15,6))
plt.bar(a.index,a.Total.values)
plt.title("Immgrants according to Area name")
plt.xlabel("Area")
plt.ylabel("Immigrants")
plt.show()
```

```python
y1=d[d['AreaName']=='Asia'].iloc[:,-35:-1].sum()
y2=d[d['AreaName']=='Africa'].iloc[:,-35:-1].sum()
y3=d[d['AreaName']=='Europe'].iloc[:,-35:-1].sum()
x=d.columns[-35:-1]
plt.figure(figsize=(20,5))
plt.plot(x,y1,label='Asia')
plt.plot(x,y2,label='Africa')
plt.plot(x,y3,label='Europe')
plt.legend()
plt.show()
```

```python
a=d.nlargest(5,columns='Total')
x=a.columns[-35:-1]
y1=a.values[0][-35:-1].astype('float')
y2=a.values[1][-35:-1].astype('float')
y3=a.values[2][-35:-1].astype('float')
y4=a.values[3][-35:-1].astype('float')
y5=a.values[4][-35:-1].astype('float')
plt.figure(figsize=(20,10))
plt.fill_between(x,y1,label=a.Country.iloc[0],alpha=0.4)
plt.fill_between(x,y2,label=a.Country.iloc[1],alpha=0.4)
plt.fill_between(x,y3,label=a.Country.iloc[2],alpha=0.4)
plt.fill_between(x,y4,label=a.Country.iloc[3],alpha=0.4)
plt.fill_between(x,y5,label=a.Country.iloc[4],alpha=0.4)
plt.legend()
plt.show()
```

```python
x=d.columns[-35:-1]
y=d.iloc[:,-35:-1].sum().values
plt.figure(figsize=(20,10))
plt.plot(x,y)
```

```python
a=d.nlargest(5,columns='Total')
x=a.columns[-35:-1]
y1=a.values[0][-35:-1].astype('float')
y2=a.values[1][-35:-1].astype('float')
y3=a.values[2][-35:-1].astype('float')
y4=a.values[3][-35:-1].astype('float')
y5=a.values[4][-35:-1].astype('float')
plt.figure(figsize=(20,10))
plt.stackplot(x,y1, y2, y3,y4,y5,labels=[a.Country.iloc[0],a.Country.iloc[1],a.Country.iloc[2],a.Country.iloc[3],a.Country.iloc[4]])
plt.legend()
plt.show()
```
