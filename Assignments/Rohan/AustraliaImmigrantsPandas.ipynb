```python
import pandas as pd
xls = pd.ExcelFile('Australia.xlsx')
d = xls.parse('Australia by Residence', skiprows=20, index_col=None)
print(d.columns)
print(d.index)
```

```python
print(d.describe())
print(d.shape)
print(d.dtypes)
print(d.iloc[1:3, 0:3])
print(d.loc[22:26,'Type':'OdName'])
```

```python
d.drop(d[d['Type']=='Emigrants'].index,inplace=True,axis=0)
```

```python
r=0.5*len(d)
d.dropna(thresh=r,axis=1,inplace=True)
```

```python
d.drop(columns=['Type','Coverage','AREA','REG','DEV'])
```

```python
d.rename(columns={'OdName':'Country','RegName':'Region'},inplace=True)
print(d)
```

```python
d.dtypes
```

```python
d.dtypes.index=d.dtypes.index.astype('str')
d.columns=d.columns.astype('str')
d.columns.map(type)
```

```python
d['Total']=d.iloc[:,-35:-1].sum(axis=1)
print(d)
```

```python
d.sort_values('Total',ascending=False)
i=d[(d.RegName=='World') & (d.Total==7968107)].index
d.drop(index=i,inplace=True)
```

```python
r=d.sort_values('Total',ascending=False)
r.tail(10)
```
