
### Titanic Plotting
- Load data generated previously from 'titanic_filtered.csv'
- plot histogram of ages of people
- plot no of people in each pclass on bar chart


- plot a horizontal bar-chart depicting the ratio of males to females
- plot a bar-chart depicting the ratio of male-survived, female-survived, male-dead, female-dead. 
- plot a stacked bar chart of number of male-survived,dead and female-survived, dead.
- plot stacked bar of survived with different pclass in same bar.


- plot box plot of ages of survived vs not-survived
- plot scatter plot of ages vs survived.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Admin\Desktop\dsc_weekday_1\Assignments\Abhishek_SG\titanic_filtered.csv")
df
```

```python
# plot histogram of ages of people
x = df.age
plt.figure(figsize = (15,6))
plt.hist(x, bins = 40,rwidth = .8)
plt.show()
plt.figure(figsize = (15,6))
df.age.plot(kind ='hist', bins = 40, rwidth = .9)
plt.show()
```

```python
a = df.pclass.value_counts()
a.plot(kind = 'bar')
plt.show()

b = df.groupby(by = "pclass").size()
b.plot(kind='bar')
plt.show()

b.plot(kind='barh')
plt.show()
```

```python
# plot a bar-chart depicting the ratio of males to females

a = df.gender.value_counts()
a.plot(kind='barh')
plt.show()
plt.barh(a.index,a)
plt.show()
```

```python
# plot a bar-chart depicting the ratio of male-survived, female-survived, male-dead, female-dead.

a = df.groupby(by = ["gender", "survived"]).size()
# plt.figure(figsize=(12,5))
x = a.plot(kind='bar')
plt.show()
x = a.plot(kind='bar', color = ['b','orange','g','r'])
plt.xticks(rotation = 0)
plt.xlabel('')
plt.show()


# x[1].set_color('r')
```

```python
# plot a stacked bar chart of number of male-survived,dead and female-survived, dead.

a = df.groupby(by = ["gender"])["survived"].value_counts()
b = a.unstack()
b.plot(kind= 'bar',stacked =True)
plt.show()

print(b)

x = [a.index[0][0],a.index[3][0]]
y = [a.values[0],a.values[2]]
y1 = [a.values[1],a.values[3]]
plt.bar(x,y1)

plt.bar(x,y, bottom = y1)
plt.show()
```

```python
a = df.groupby(by = ['pclass'])['survived'].value_counts()
b = a.unstack()
b.plot(kind ='bar', stacked =True)
plt.show()
c = df.groupby(by = ['survived'])['pclass'].value_counts()
d = c.unstack()
d.plot(kind ='bar', stacked =True)
plt.show()
```

```python
# plot box plot of ages of survived vs not-survived

a = df.groupby(by = 'survived').get_group(0).age
plt.boxplot([a])
plt.show()
b = df.groupby(by = 'survived').get_group(1).age
plt.boxplot([b])
plt.show()
```

```python
# plot scatter plot of ages vs survived.
plt.figure(figsize=(15,6))
plt.scatter(df.age, df.survived)
plt.show()
```
