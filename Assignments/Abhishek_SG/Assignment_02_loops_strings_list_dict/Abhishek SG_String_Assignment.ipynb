```
#W A P which applies strip() method if any string, which will be taken from user, starts and ends with space, or applies 
#rstrip() method if that string only ends with space or applies lstrip() method if that string only starts with a space.

#For example:-
#input:- '    Python   '
#output:- 'Python'
ip = '    Python   '
if ip.startswith(' ') and ip.endswith(' ') == True:
  print(ip.strip())
  print("Used strip()")
elif ip.startswith(' ') == True:
  print(ip.lstrip())
  print("Used lstrip")
else:
  print(ip.rstrip())
  print("Used rstrip")

#input:- '    Python'
#output:- 'Python'

ip1= '    Python'
if ip1.startswith(' ') and ip1.endswith(' ') == True:
  print(ip1.strip())
  print("Used strip()")
elif ip1.startswith(' ') == True:
  print(ip1.lstrip())
  print("Used lstrip")
else:
  print(ip1.rstrip())
  print("Used rstrip")


#input:- 'Python   '
#output:- 'Python'

ip2 ='Python   '

if ip2.startswith(' ') and ip2.endswith(' ') == True:
  print(ip2.strip())
  print("Used strip()")
elif ip2.startswith(' ') == True:
  print(ip2.lstrip())
  print("Used lstrip")
else:
  print(ip2.rstrip())
  print("Used rstrip")
```

```
#Write a Python program to find the total number of times letter 'p' is appeared in the below string:-
    
#Input:- 'peter piper picked a peck of pickled peppers.'
#Output:- 9

ip = 'peter piper picked a peck of pickled peppers.'
count = 0
i=0
while i < len(ip):
  a = ip.find('p', i , len(ip))
  if a >= 0:
    count+=1
    i=a+1
  else:
    break
print(count)

```
