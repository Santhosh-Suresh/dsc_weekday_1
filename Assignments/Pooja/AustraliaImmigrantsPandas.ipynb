```python
import pandas as pd
```

```python
# Read sheet "Australia by Residence" from Australia.xlsx
# Ignore the first 20 rows.

a = pd.read_excel('../Pooja/Australia.xlsx',sheet_name = 1,skiprows = 20)
a
```

```python
# List all Column and Row indexes(Print their names)
print('column index =',a.columns)
print('row index = ',a.index)
# print(a.index.values)
```

```python
# Describe the dataset
a.describe(include = 'all')
```

```python
# Use values and index attributes on both Row and Column Indexes
print(a.columns.values)
print()
print(a.index.values)
```

```python
# Check Shape, Datatypes and other attributes and observe the datatype of each columns' data
print(a.shape)
print()
print(a.dtypes)
print()
print(a.ndim)
```

```python
# Delete all Data where the Type = "Emigrants" or Type is not "Immigrants" (Row Deletion)/
a.drop(a[a['Type'] == 'Emigrants'].index , inplace = True)
a.shape
```

```python
# Find columns for which more than 50% data is missing and drop them (Column Deletion)
perc = len(a) * .5
a.dropna(thresh = perc, axis = 1, inplace = True)
```

```python
# Drop Unnessary Columns: Type, Coverage, Area, Reg, Dev (Column deletion) Your columns should now look like 
# # "OdName,AreaName,RegName,DevName,1980 ... 2008"
# print(f'Columns before deleting the unnecessary column are: \n{a.columns}')
# # a.drop(columns = ['Type', 'Coverage', 'AREA', 'REG', 'DEV'],axis = 1, inplace = True)
print(f'Columns after deleting the unnecessary column are: \n{a.columns}')
```

```python
# Rename columns :OdName -> Country; RegName -> Region
a.rename(columns={"OdName" : "Country", 'RegName' : "Region"}, inplace=True)
a.columns
```

```python
# Convert Column Names with Int type to Strings to avoid index errors with string indexing "1980, 1981
# etc." => " '1980', '1981' ... " (Use rename)
for i in a.columns:
    if type(i) == int:
        a.rename(columns ={i:f'{i}'}, inplace = True )
a.columns
```

```python
# Find Total number of immigrants for all years, and create a new column Total in data Frame
a['Total'] = a.iloc[:, 4:].sum(axis = 1)
a.columns
```

```python
print(a.head(2))
```

```python
# Sort Data by the new column Total in descending order and analyze the fist few rows. (get first few rows using head)
a_sorted = a.sort_values('Total', ascending = False)
a_sorted.head()
```

```python
# find something weird in the dataset ? Delete that row 'world'
a_sorted.drop(a_sorted[a_sorted['AreaName'] == 'World'].index, inplace=True)
a_sorted.head()
```

```python
# Analyze last few rows also in the sorted result (Don't sort inplace)
a_sorted.tail()
```

```python
# Check if these have all unique values or not:

# countries

print(len(a.Country.unique()) == a.shape[0])
print(a.shape[0])




```

```python
# continents (AreaName)
print(a.AreaName.unique())
print(len(a.AreaName.unique()) == a.shape[0])
```

```python
# country continent pairs
print(len(a.groupby(['Country', 'AreaName']).groups))
print(len(a.groupby(['Country', 'AreaName']).groups) == a.shape[0])
```

```python
# Find countries for which no of immigrants for all years is 0
a.loc[a.Total == 0 , ['Country']]
```

```python
# Find count of above countries
a.loc[a.Total == 0 , ['Country']].count()
```

```python
# Find countries for which no of immigrants is 0 for any year
a.loc[a[a.columns[4:]].apply(lambda x:x==0).any(axis=1),['Country']]
```

```python
# Drop the row where AreaName is 'World'
print(a.shape)
a.drop(a[a['AreaName'] == 'World'].index, inplace= True)
print(a.shape)
```

```python
# Save the dataset to a new file AustraliaFiltered.xlsx
a.to_excel('AustraliaFiltered.xlsx', index = False)
```
