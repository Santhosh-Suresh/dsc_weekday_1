```python
import pandas as pd

# Read sheet "Australia by Residence" from Australia.xlsx
# Ignore the first 20 rows.

df=pd.read_excel('Australia.xlsx',sheet_name=1,header=20)
df
```

```python
#Ignore the first 20 rows.
#skip rows can be used here; which was done above already
```

```python
#List all Column and Row indexes(Print their names)
print(df.index)

print()
print(df.columns)
```

```python
#Describe the dataset
df.describe()
```

```python
#Use values and index attributes on both Row and Column Indexes
print(df.index)
print()
print(df.loc[1])
```

```python
#Check Shape, Datatypes and other attributes and observe the datatype of each columns' data
print(df.shape)
print()
print(df.dtypes)
```

```python
# Drop
# Delete all Data where the Type = "Emigrants" or Type is not "Immigrants" (Row Deletion)
df.drop(df[df.Type!='Immigrants'].index,axis=0,inplace=True)
df
```

```python
# Find columns for which more than 50% data is missing and drop them (Column Deletion)
#df.dropna(axis = 'columns', thresh = df.shape[0]*0.50)

#print(df.head(2))
# df.isna(any
        
a=len(df)*.5
print(a)
df.dropna( thresh = a,axis = 'columns',inplace=True)
df
#print(df.count()/224*100)
#df.columns(0.50)
```

```python
# Drop Unnessary Columns: Type, Coverage, Area, Reg, Dev (Column deletion) Your columns should now look like "OdName,AreaName,RegName,DevName,1980 ... 2008"
df.drop(columns=['Type', 'Coverage', 'AREA', 'REG', 'DEV'],inplace=True)
print()
print('After Dropping the columns')
df.columns
```

```python
# Rename columns :OdName -> Country; RegName -> Region

df.rename(columns={'OdName':'Country','RegName':'Region'},inplace=True)
df.columns
```

```python
# Convert Column Names with Int type to Strings to avoid index errors with string indexing "1980, 1981 etc." => " '1980', '1981' ... " (Use rename)
for i in df.columns:
    if type(i)==int:
        df.rename(columns={i:'%s' %i}, inplace=True)
        




```

```python
# Find Total number of immigrants for all years, and create a new column Total in data Frame
#iloc- integer location(rows,integer colums)- [finding the integer colum]
df['Total'] = df.iloc[:, 4:].sum(axis = 1)
df.columns
df.head()
```

```python
# Sort Data by the new column Total in descending order and analyze the fist few rows. (get first few rows using head)
df1=df.sort_values(by='Total',ascending=False)
df1.head()
```

```python
#find something weird in the dataset ? Delete that row 'world'

df1.drop(df1[df1['AreaName'] == 'World'].index, inplace=True)
df1.head()
```

```python
#Analyze last few rows also in the sorted result (Don't sort inplace)
# df1.tail()
```

```python
# Check if these have all unique values or not:

# countries
# continents (AreaName)
# country continent pairs

if len(df.Country.unique()) == len(df.Country):
    print("Country : all unique values present")
else:
    print("Country : values are not unique")
    
if len(df.AreaName.unique()) == len(df.AreaName):
    print("continents: all unique values present")
else:
    print("continents: values are not unique")

df["temp"] = df["Country"]+df["AreaName"]

if len(df["temp"].unique()) == len(df["temp"]):
    print("Country,continents pair: all unique values present")
else:
    print("Country,continents pair: values are not unique")
```

```python
# Find countries for which no of immigrants for all years is 0



df["Country"][df.Total==0]
```

```python
# Find count of above countries
df["Country"][df.Total==0].count()
```

```python
# Find countries for which no of immigrants is 0 for any year
df.Country[df[df.columns[4:]].apply(lambda x:x==0, axis =0).any(axis=1)]
```

```python
# Save Data

# Drop the row where AreaName is 'World'


df.drop(df[df['AreaName'] == 'World'].index, axis = 0, inplace= True)
```

```python
# Save the dataset to a new file AustraliaFiltered.xlsx
df.to_excel('AustraliaFiltered.xlsx', index = False)
```
