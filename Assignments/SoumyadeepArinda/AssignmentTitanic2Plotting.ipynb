
### Titanic Plotting
- Load data generated previously from 'titanic_filtered.csv'
- plot histogram of ages of people
- plot no of people in each pclass on bar chart


- plot a horizontal bar-chart depicting the ratio of males to females
- plot a bar-chart depicting the ratio of male-survived, female-survived, male-dead, female-dead. 
- plot a stacked bar chart of number of male-survived,dead and female-survived, dead.
- plot stacked bar of survived with different pclass in same bar.


- plot box plot of ages of survived vs not-survived
- plot scatter plot of ages vs survived.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("titanic_filtered.csv")
```

```python
# plot histogram of ages of people
df.age.plot(kind = 'hist')
```

```python
# plot no of people in each pclass on bar chart
a = df.groupby("pclass").size()
print(a)
print()
a.plot(kind = "bar")
```

```python
# plot a bar-chart depicting the ratio of males to females
a = df.gender.value_counts(normalize = True) * 100
a.plot(kind = 'barh')
```

```python
# plot a bar-chart depicting the ratio of male-survived, female-survived, male-dead, female-dead.
a  = df.groupby(by = ['gender', 'survived']).size()
a.plot(kind = 'bar', color = 'c')
```

```python
# plot a stacked bar chart of number of male-survived,dead and female-survived, dead.
a = df.groupby(['gender','survived'])
print(a.size())

plot = a.size().unstack()

print(plot)
plot.plot(kind = 'bar', stacked = True)
plt.show()
```

```python
# plot stacked bar of survived with different pclass in same bar
a = df.groupby(by = ['pclass','survived']).size().unstack()
print(a)
a.plot(kind='bar', stacked=True)


b = df.groupby(by = ['survived', 'pclass']).size().unstack()
print(b)
b.plot(kind='bar', stacked=True)
```

```python
# plot box plot of ages of survived vs not-survived
grp = df.groupby('survived')
survived_age = grp.get_group(0).age
non_survived_age = grp.get_group(1).age

survived_age.plot(kind = 'box', label = 'ages of survivers')
plt.show()
non_survived_age.plot(kind = 'box', label = 'ages of non-survivers')
plt.show()
```

```python
# plot scatter plot of ages vs survived.
plt.figure(figsize=(12,5))
plt.scatter(df.age, df.survived, alpha = 0.4)
plt.show()
```
