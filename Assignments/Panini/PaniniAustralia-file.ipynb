```python
# Read sheet "Australia by Residence" from Australia.xlsx
# Ignore the first 20 rows.
# List all Column and Row indexes(Print their names)

import pandas as pd
import numpy as np
anz = pd.read_excel('Australia.xlsx',sheet_name=1,skiprows=20)

print('Row Index=       ', anz.index)
print('Column Index=    ',anz.columns)
```

```python
#     Check Shape, Datatypes and other attributes and observe the datatype of each columns' data
print(anz.shape)
# datatype of each columns data
for i in Col_attr:
    print(anz[i].dtype)
# dimension
print(anz.ndim)
```

# Drop

    Delete all Data where the Type = "Emigrants" or Type is not "Immigrants" (Row Deletion)
    Find columns for which more than 50% data is missing and drop them (Column Deletion)
    Drop Unnessary Columns: Type, Coverage, Area, Reg, Dev (Column deletion) Your columns should now look like "OdName,AreaName,RegName,DevName,1980 ... 2008"


```python
#Delete all Data where the Type = "Emigrants" or Type is not "Immigrants" (Row Deletion)
anz_wip = anz.copy()
print(anz_wip['Type'].value_counts())
anz_wip = anz_wip.drop(anz_wip[anz_wip['Type']=='Emigrants'].index)
#df.drop(df.index[df['line_race'] == 0], inplace = True)
#df = df[df.line_race != 0]
```

```python
anz_wip
```

```python
anz_wip.shape[0]
```

```python
#Find columns for which more than 50% data is missing and drop them (Column Deletion)
anz_wip.replace('..', np.nan, inplace = True)
anz_wip.replace(0, np.nan,inplace = True)
# per = 0.5 *len(anz_wip)
# anz_wip.dropna(axis=0,thresh = per)
anz_wip.drop(anz_wip.columns[anz_wip.isna().mean()>0.5], axis = 1, inplace = True)
```

```python
anz_wip
```

```python
#Drop Unnessary Columns: Type, Coverage, Area, Reg, Dev (Column deletion) Your columns should now look 
#like "OdName,AreaName,RegName,DevName,1980 ... 2008"
cols = ['Type', 'Coverage', 'AREA', 'REG', 'DEV']
anz_wip.drop(cols, axis=1, inplace = True)
# anz_wip.drop(anz_wip.columns[anz_wip.isna().mean()>0.5], axis = 1, inplace = True)
```

```python
anz_wip
```

# Rename

    Rename columns :OdName -> Country; RegName -> Region
    Convert Column Names with Int type to Strings to avoid index errors with string indexing "1980, 1981 etc." => " '1980', '1981' ... " (Use rename)
    Find Total number of immigrants for all years, and create a new column Total in data Frame


```python
# Rename columns :OdName -> Country; RegName -> Region
anz_wip = anz_wip.rename(columns = {'OdName':'Country','RegName':'Region'  })
z= [x for x in anz_wip.columns if type(x)==int]
z_str =map(str, z)
p = {x:y for x,y in zip(z, list(z_str))}
anz_wip = anz_wip.rename(columns = p)
anz_wip
```

```python
# Find Total number of immigrants for all years, and create a new column Total in data Frame
anz_wip['Total'] = anz_wip.loc[:, '1980':].sum(axis=1)
anz_wip
```
