```python
import pandas as pd
```

```python
df = pd.read_csv('titanic_dataset.csv')
#df = pd.read_csv(r'https://raw.githubusercontent.com/leangaurav/dsc_weekday_1/master/Assignments/titanic_dataset.csv')
```

```python
df.describe(include='all')
```

```python
#Find columns with missingdata
df_missing= df.isna().sum(axis=0)
# r = df.count()
# r[r<df.shape[0]].index
# df.columns[titanic.isnull().any()]
print(df_missing)
```

```python
#Print shape of dataset
df.shape
```

```python
#drop columns with more than 25% missing data
# df.dropna(thresh=int(0.75*df.shape[0]))
perc= 0.75*df.shape[0]
clean_df=df.dropna(thresh=perc, axis = 1, )
print(clean_df.columns)
```

```python
#drop columns having independent values(which do not affect the survival rate)
indp_cols = ['name','Unnamed: 0' ]
clean_df.drop(indp_cols, inplace=True, axis = 1)
```

```python
clean_df
```

```python
#check data types of all columns
clean_df.dtypes
```

```python
#convert price to numeric
# fare_clean1= list(clean_df['fare'])
# # fare_num1 = [ float(s[1:]) if type(s)== str else s if type(s)!= str for s in fare_clean1]
# fare_num1 = [ float(s[1:]) for s in fare_clean1 if type(s)== str else s if type(s)!= str ]
# clean_df[clean_df['fare']!=np.nan]
# convert price to numeric
clean_df['fare'] = clean_df.fare.str.replace('$' , '')
clean_df.fare = clean_df.fare.astype(float)
clean_df['fare']
```

```python
#find columns still having missing/na values and also count of missing data
print(clean_df.isna().any())
print(clean_df.isna().sum())
```

```python
#fill na with mean for fare and age column column
clean_df['age'].fillna(clean_df['age'].mean(), inplace=True)
clean_df['fare'].fillna(clean_df['fare'].mean(), inplace=True)
print(clean_df.isna().any())
```

```python
#drop na values for embarked column.
clean_df.dropna(axis=0,subset=['embarked'], inplace=True)
clean_df.shape
```

```python
#dump the dataframe to a csv file 'titanic_filtered.csv'.
clean_df.to_csv(r'Panini_titanic_filtered.csv')
```

```python
#for surviced column replace 0 with D and 1 with A
# clean_df['survived'].replace([0, 1],['D', 'A'], inplace = True)
clean_df[clean_df['survived']==0]='D'
clean_df[clean_df['survived']==1]='A'
clean_df
```

```python
#find the frequency of different values in survived column
clean_df['gender'].value_counts()
```

```python
#group by gender and survived and see the counts in each category

grp=clean_df.groupby(by=['gender','survived'])
grp.count()
```

```python
#find different pclass and no of people in each class
grp1 = clean_df.groupby(['pclass']).size()
grp1
```

```python
#find top 5 people with highest values of age. Count no of male and females in the top 5
grp2=clean_df.groupby(by = ['age','gender']).size()
grp2
```

```python
#find max age male and female who survived
clean_df[clean_df['survived']=='A'].groupby(['gender']).max()
```
