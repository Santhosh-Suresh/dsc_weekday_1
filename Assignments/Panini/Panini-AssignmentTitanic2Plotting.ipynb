
### Titanic Plotting
- Load data generated previously from 'titanic_filtered.csv'
- plot histogram of ages of people
- plot no of people in each pclass on bar chart


- plot a horizontal bar-chart depicting the ratio of males to females
- plot a bar-chart depicting the ratio of male-survived, female-survived, male-dead, female-dead. 
- plot a stacked bar chart of number of male-survived,dead and female-survived, dead.
- plot stacked bar of survived with different pclass in same bar.


- plot box plot of ages of survived vs not-survived
- plot scatter plot of ages vs survived.

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('Panini_titanic_filtered.csv')
```

```python
# plot histogram of ages of people
a=df.age
plt.figure(figsize=(20,5))
r =plt.hist(a, bins = 30, color = 'c', label = 'age' , rwidth=.9)
plt.title('Age')
plt.xticks(r[1])
print(r[0])
print(r[1])
plt.show()
```

```python
#plot no of people in each pclass on bar chart
ac=df.pclass.value_counts()
print((ac))
plt.xticks(ac.index)
plt.bar(ac.index, ac.values)
plt.title('no. of people by Passenger Class')
plt.xlabel('Passenger Class')
plt.show()
```

```python
# plot a bar-chart depicting the ratio of males to females
bd = df.gender.value_counts()
plt.barh(bd.index, bd.values,  align='center')
plt.title('Gender Ratio')
plt.xlabel('no. by Gender')
plt.ylabel('Gender')
plt.show()
print(bd )
```

```python
# plot a bar-chart depicting the ratio of male-survived, female-survived, male-dead, female-dead.
bc=df[['gender', 'survived']]
dc=df.groupby(['gender', 'survived'])
cd=dc.pclass.count()
cd.plot.bar()
plt.show()
```

```python
# plot a stacked bar chart of number of male-survived,dead and female-survived, dead
df_D=df[df['survived']=='D']
df_A=df[df['survived']=='A']
dd=df_D['gender'].value_counts()
aa=df_A['gender'].value_counts()
aa.sort_values(ascending=True, inplace=True)
plt.bar(dd.index, dd.values)
plt.bar(aa.index,aa.values, bottom = dd.values )
plt.legend(['D', 'A'])
plt.title('Mortality rate by gender')
plt.xlabel('Gender')
plt.ylabel('Mortality ratio')
plt.show()
```

```python
# plot stacked bar of survived with different pclass in same bar
df_D=df[df['survived']=='D']
df_A=df[df['survived']=='A']
dd=df_D['pclass'].value_counts()
aa=df_A['pclass'].value_counts()
aa.sort_index(ascending=True, inplace=True)
dd.sort_index(ascending=True, inplace=True)
plt.bar(dd.index, dd.values)
plt.bar(aa.index,aa.values, bottom = dd.values )
plt.legend(['D', 'A'])
plt.title('Mortality rate by pclass')
plt.xlabel('pclass')
plt.xticks(dd.index)
plt.ylabel('Mortality ratio')
plt.show()
```

```python
# plot box plot of ages of survived vs not-survived

df_D=df[df['survived']=='D']
df_A=df[df['survived']=='A']
dd=df_D['age']
aa=df_A['age']
plt.subplot(1,2,1)
plt.boxplot([dd])
plt.xlabel('D')
plt.subplot(1,2,2)
plt.boxplot(aa)
plt.xlabel('A')
plt.show()
```

```python
# plot scatter plot of ages vs survived.
df_age=df['age']
df_survived=df['survived']
plt.scatter(df_age, df_survived, alpha=0.1)
plt.show()
```
