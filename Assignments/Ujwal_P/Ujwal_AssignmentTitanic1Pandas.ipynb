```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("titanic_dataset_copy.csv")
print(df.head())
```

```python
# Read data and describe it
df.describe(include='all')
```

```python
# Find columns with missingdata
# give me names of columns which are missing alteast one value.
# print(df.isna().any())
print(df.columns[df.isna().any()])

# or another way

r = df.count()
print(r)
i=r < df.shape[0]
print(i)
r[i].index
```

```python
# Print shape of dataset
print(df.shape)
```

```python
# drop columns with more than 25% missing data
print(df.count()/1309 * 100)

print()

d=df.dropna(thresh=len(df)*0.75,axis='columns')
print(d.columns)
print(df.columns)

df.dropna(thresh=len(df)*0.75,axis='columns',inplace=True)  # to do it in original dataframe
print(df.columns)
```

```python
# drop columns having independent values(which do not affect the survival rate).
# the columns which are of above type are- 'Unnamed: 0', 'name', 'boat','body'
# out of above only 'Unnamed: 0', 'name' needs to be removed as remaining are already removed
df.drop(columns=['Unnamed: 0','name'],inplace=True)
print(df.columns)
```

```python
# check data types of all columns
print(df.dtypes)
```

```python
# convert price to numeric
print(df['fare'])
df['fare'] = df['fare'].replace('\$','',regex=True).astype(float)
print(df['fare'])
print(df.dtypes)
```

```python
# find columns still having missing/na values and also count of missing data
arr=df.columns[df.isna().any()]
print(arr)
print()
print(df[arr].count())
print()
print(df[arr].isna().sum())
```

```python
# fill na with mean for fare and age column column.
print(df[['age','fare']].mean())
print()
d=df[['age','fare']].fillna(df.mean(),inplace=False)
print(d[['age','fare']].isna().sum())
print()
print(df)
print()
print(d)

df.fillna(df[['age','fare']].mean(),inplace=True)
print(df)

# or
# f[['age','fare']]=df[['age','fare']].fillna(df.mean(),inplace=False)
```

```python
# drop na values for embarked column.

d=df.dropna()
print(d)
df.dropna(inplace=True)
```

```python
# dump the dataframe to a csv file 'titanic_filtered.csv'.
df.to_csv('titanic_filtered.csv',index=False)
```

```python
# for survived column replace 0 with D and 1 with A
d=df['survived'].replace([0,1],['D','A'])
print(d)
df['survived'].replace([0,1],['D','A'],inplace=True)
print(df)
print(df.dtypes)
# dtype of survived columns converted from int64 to object
```

```python
# find the frequency of different values in survived column
print(df['survived'].unique())
print()
print(df['survived'].value_counts())
```

```python
# group by gender and survived and see the counts in each category
grp=df.groupby(by=['gender','survived'])
print(grp)
print()
print(grp.groups)
print()
print(grp.embarked.count()) # any column name can be used for counting - embarked etc
```

```python
# find different pclass and no of people in each class
print(df['pclass'].unique())
print()
print(df['pclass'].value_counts())
```

```python
# find top 5 people with highest values of age. Count no of male and females in the top 5
print(df.nlargest(5,['age']))
d1=pd.DataFrame(df.nlargest(5,['age']))
print()
print(d1['gender'].value_counts())
```

```python
# find max age male and female who survived
# print(grp['age'].max())

# print(df[df.survived=='A'])
# print(df[["gender","age"]])
# print(df[["gender","age"]][df.survived=='A'])


# df[df['survived'] == 'A'].groupby('gender')['age'].max()
# df[df['survived'] == 'A'].groupby('gender').age.max()
df[["gender","age"]][df.survived =="A"].groupby('gender').max()
```

```python
# get average age by gender
# df.groupby('gender')['age'].mean()
df[["gender","age"]].groupby('gender').mean()
```
