```python
import pandas as pd 
import numpy as np
```

```python
# Read sheet "Australia by Residence" from Australia.xlsx
# Ignore the first 20 rows.
# List all Column and Row indexes(Print their names)

# df=pd.read_excel("Australia.xlsx",sheet_name="Australia by Residence")
df=pd.read_excel("Australia.xlsx",sheet_name="Australia by Residence",skiprows=list(range(0,20)))
print(df.head(5))
print()

print(df.columns)
print()

print(df.index)
```

# Dataset Analysis

```python
# Describe the dataset

df.describe()
```

```python
# Use values and index attributes on both Row and Column Indexes
# print(df.columns.values)
# print()
print(df["AREA"].index)
print()
print(df["AREA"].values)
```

```python
print(df.index)
print()
print(df.loc[45])
```

```python
# Check Shape, Datatypes and other attributes and observe the datatype of each columns' data
print(df.shape)
print()
print(df.dtypes)
print()
print(df.size)
print()
# print(df.T)
```

# Data

```python
# Delete all Data where the Type = "Emigrants" or Type is not "Immigrants" (Row Deletion)
# df[df.Type=="Emigrants"].index
df.drop(df[df.Type=="Emigrants"].index,axis=0,inplace=True)
df
```

```python
# Find columns for which more than 50% data is missing and drop them (Column Deletion)
df.replace("..",np.nan,inplace=True)
print(df.count())
df.dropna(axis=1,thresh=0.5*len(df),inplace=True)
df
```

```python
# Drop Unnessary Columns: Type, Coverage, Area, Reg, Dev (Column deletion) Your columns should 
# now look like "OdName,AreaName,RegName,DevName,1980 ... 2008"
df.drop(["Type","Coverage","AREA","REG","DEV"],axis=1,inplace=True)
df.columns
```

# Rename

```python
# Rename columns :OdName -> Country; RegName -> Region
df.rename({"OdName":"Country","RegName":"Region"},axis=1,inplace=True)
```

```python
# Convert Column Names with Int type to Strings to avoid index errors with string indexing 
#"1980, 1981 etc." => " '1980', '1981' ... " (Use rename)
arr=df.columns.values
for elem in arr:
    if type(elem)==int:
#         print(elem)
        df.rename({elem:str(elem)},axis=1,inplace=True)
    
df.columns
```

```python
# Find Total number of immigrants for all years, and create a new column Total in data Frame
arr=df.loc[:,'1980':'2008'].sum(axis=1)
df.insert(len(df.columns),'Total',arr,allow_duplicates=True)
df
```

# Head Tail Sort

```python
# - Sort Data by the new column Total in descending order and analyze the fist few rows. (get first few rows using head)
#     - find something weird in the dataset ? Delete that row 'world'
# - Analyze last few rows also in the sorted result (Don't sort inplace)
df_sorted_by_total=df.sort_values('Total',axis=0,ascending=False)

df_sorted_by_total

```

```python
df.drop(447,axis=0,inplace=True)
df
```

```python
# - Check if these have all unique values or not:
#     - countries
#     - continents (AreaName)
#     - country continent pairs
print(df['Country'].unique().size==df['Country'].count())
print(df['AreaName'].unique().size==df['AreaName'].count())
ans=df.groupby(['Country','AreaName'],axis=0)
print(len(ans.groups)==df.index.size)
```

```python
# - Find countries for which no of immigrants for all years is 0
# - Find count of above countries

# method 1
print(df[['Country']][df['Total']==0].values)
print()
print(df[['Country']][df['Total']==0].values.size)
print()

# method 2
print(df['Country'][(df.loc[:,'1980':'2008']==0).all(axis=1)].values)
print()
print((df.loc[:,'1980':'2008']==0).all(axis=1).sum())
```

```python
# - Find countries for which no of immigrants is 0 for any year

print(df['Country'][(df.loc[:,'1980':'2008']==0).any(axis=1)].values)
print()
print((df.loc[:,'1980':'2008']==0).any(axis=1).sum())
```

# Save Data

```python
# - Drop the row where AreaName is 'World'-


df.drop(df[df.AreaName=='World'].index.values,inplace=True)
```
