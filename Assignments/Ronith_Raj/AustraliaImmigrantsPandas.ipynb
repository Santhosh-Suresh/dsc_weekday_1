```python
import pandas as pd
import numpy as np
```

```python
australia = pd.read_excel('Australia.xlsx','Australia by Residence', skiprows=range(1, 20),header=1)
```

```python
print(list(australia.index))
```

```python
(australia.describe(include = 'all'))
```

```python
australia.index.values
```

```python
# Check Shape, Datatypes and other attributes and observe the datatype of each columns' data

print(australia.shape)
print(australia.size)
print(australia.dtypes)
```

```python
# Delete all Data where the Type = "Emigrants" or Type is not "Immigrants" (Row Deletion)
australia.drop(australia[australia['Type'] == 'Emigrants'].index , inplace = True)
```

```python
print(australia.shape)
```

```python
# Find columns for which more than 50% data is missing and drop them (Column Deletion)
per_50 = len(australia) * .5
australia.dropna(thresh = per_50, axis = 1, inplace = True)
```

```python
# Drop Unnessary Columns: Type, Coverage, Area, Reg, Dev (Column deletion) Your columns should now look like "OdName,AreaName,RegName,DevName,1980 ... 2008"
print(f'Columns before deleting the unnecessary column are: \n{australia.columns}')

australia.drop(columns = ['Type', 'Coverage', 'AREA', 'REG', 'DEV'],axis = 1, inplace = True)

print(f'Columns after deleting the unnecessary column are: \n{australia.columns}')
```

```python
# Rename columns :OdName -> Country; RegName -> Region
australia.rename(columns={"OdName" : "Country", 'RegName' : "Region"}, inplace=True)
australia.columns
```

```python
# Convert Column Names with Int type to Strings to avoid index errors with string indexing "1980, 1981 etc." => " '1980', '1981' ... " 
# (Use rename)
# cols = australia.columns[4:]
for i in australia.columns:
    if type(i) == int:
        australia.rename(columns ={i:f'{i}'}, inplace = True )
```

```python
# Find Total number of immigrants for all years, and create a new column Total in data Frame
australia['Total'] = australia.iloc[:, 4:].sum(axis = 1)
```

```python
print(australia.columns)
print(australia.head(2))
```

```python
# Sort Data by the new column Total in descending order and analyze the fist few rows. (get first few rows using head)
australia_sorted = australia.sort_values('Total', ascending = False)
australia_sorted.head()
```

```python
# australia[australia['AreaName'] == 'World']
```

```python
# find something weird in the dataset ? Delete that row 'world'
australia_sorted.drop(australia_sorted[australia_sorted['AreaName'] == 'World'].index, inplace=True)
australia_sorted.head()
```

```python
# Analyze last few rows also in the sorted result (Don't sort inplace)

australia_sorted.tail()
```

```python
# Check if these have all unique values or not:

# countries
print(len(australia.Country.unique()) == australia.shape[0])
print(australia.shape[0])

# continents (AreaName)
print(australia.AreaName.unique())
print(len(australia.AreaName.unique()) == australia.shape[0])
```

```python
# country continent pairs
print(len(australia.groupby(['Country', 'AreaName']).groups))
print(len(australia.groupby(['Country', 'AreaName']).groups) == australia.shape[0])
```

```python
# Find countries for which no of immigrants for all years is 0
australia.loc[australia.Total == 0 , ['Country']]
```

```python
# Find count of above countries
australia.loc[australia.Total == 0 , ['Country']].count()
```

```python
# Find countries for which no of immigrants is 0 for any year
australia.loc[australia[australia.columns[4:]].apply(lambda x:x==0).any(axis=1),['Country']]
```

```python
# Drop the row where AreaName is 'World'
print(australia.shape)
australia.drop(australia[australia['AreaName'] == 'World'].index, inplace= True)
print(australia.shape)
```

```python
australia.head(1)
australia.drop(columns=['2009','2010','2011','2012','2013'] , inplace=True)
```

```python
# Save the dataset to a new file AustraliaFiltered.xlsx
australia.to_excel('AustraliaFiltered.xlsx', index = False)
```
